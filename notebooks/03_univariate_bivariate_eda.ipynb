```python
import numpy as np;
import pandas as pd;
import pandas as pd

url = "https://raw.githubusercontent.com/irfan-mohamed/eda-project/main/data/interim/cleaned_day2.csv"

df = pd.read_csv(url)
df.head()
```

```python
import matplotlib.pyplot as plt;
import seaborn as sns;

sns.set_theme(style='whitegrid', palette='pastel')
```

#UNIVARIATE ANALYSIS
##Selling Price Distribution (Histogram)

```python
plt.figure(figsize = (8,5))
sns.histplot(df['selling_price'], kde = True)
plt.xscale('log')
plt.title("Distribution of selling price")
plt.xlabel("Selling price")
plt.ylabel("Frequency")
plt.show()
plt.savefig('selling_price_distribution.png',dpi=500,bbox_inches='tight')
```

- The selling price is right skewed. which means the most of the car prices are varies from low to mid. there is a less amount of luxury cars.

##Vehicle driven distribution (Histogram)

```python
plt.figure(figsize=(8,5))
sns.histplot(df['km_driven'], kde=True)
plt.xscale('log')
plt.title("Distribution of Kilometers Driven")
plt.xlabel("Kilometers Driven")
plt.ylabel("Frequency")
plt.show()
plt.savefig('kilometers_driven_distribution.png',dpi=500,bbox_inches='tight')
```

- The Kilometers driven distribution also in positively skewed. Most of the vehicles driven a range of kilometers from low to mid level. there is extreme kilometers driven vehicles also there.

##Manufacturing year distribution (Histogram)

```python
plt.figure(figsize=(8,5))
sns.histplot(df['year'], bins=30, kde = True)
plt.title("Distribution of Manufacturing Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
plt.savefig('manufacturing_year_distribution.png',dpi=500,bbox_inches='tight')
```

- It shows most of the vehicles are from the manufacturing year of 2010. This is a negatively skewed distribution.

##Fuel type distribution (Count Plot)

```python
plt.figure(figsize=(7,4))
sns.countplot(x='fuel', data=df)
plt.title("Fuel Type Distribution")
plt.xlabel("Fuel Type")
plt.ylabel("Count")
plt.show()
plt.savefig('Fuel_type_distribution.png',dpi=500,bbox_inches='tight')
```

- The most of the market is dominated by Diesel and Petrol vehicles. There is less LPG and CNG vehicles are there.

## Transmission Distribution (Count Plot)

```python
plt.figure(figsize=(6,4))
sns.countplot(x='transmission', data=df)
plt.title("Transmission Type Distribution")
plt.xlabel("Transmission")
plt.ylabel("Count")
plt.show()
plt.savefig('transmission_type_distribution.png',dpi=500,bbox_inches='tight')
```

 - Manual transmission vehicles significantly outnumber automatic cars.
 so, it defines the cost sensitivity of the market.

## Ownership distribution (Count Plot)

```python
plt.figure(figsize=(6,4))
sns.countplot(x='owner', data=df)
plt.title("Owner Category Distribution")
plt.xlabel("Owner Category")
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.show()
plt.savefig('owner_category_distribution.png',dpi=500,bbox_inches='tight')
```

```python
plt.figure(figsize=(6,5))
sns.histplot(df['mileage_km_per_ltr'], bins=30, kde = True)
plt.title("Mileage Distribution")
plt.xlabel("Mileage (km/ltr)")
plt.ylabel("Frequency")
plt.show()
plt.savefig('mileage_distribution.png',dpi=500,bbox_inches='tight')
```

## Engine value distribution (Histogram)

# Bivariate Analysis

##Selling Price vs Kilometers Driven

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='km_driven', y='selling_price', data=df)
plt.title("Selling Price vs Kilometers Driven")
plt.xlabel("Kilometers Driven")
plt.ylabel("Selling Price")
plt.show()
plt.savefig('selling_price_vs_kilometer_driven.png',dpi=500,bbox_inches='tight')
```

- Here we can see a negative relationship. higher kilometer shows lower selling price.

## Selling Price vs Manufacturing Year

```python
plt.figure(figsize = (8,5))
sns.scatterplot(x='year', y = 'selling_price', data = df)
plt.title('Selling Price vs Manufacturing Year')
plt.xlabel('Manufacturing Year')
plt.ylabel('Selling Price')
plt.show()
plt.savefig('selling_price_vs_manufacturing_year.png',dpi=500,bbox_inches='tight')
```

- This Figure shows that the new vehicles demand more price.

## Selling Price vs Fuel Type

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = 'fuel', y = 'selling_price', data = df)
plt.title('Selling Price vs Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Selling Price')
plt.show()
plt.savefig('selling_price_vs_fuel_type.png',dpi=500,bbox_inches='tight')
```

- LPG vehicles and CNG vehicles shows lower selling prices.

## Selling Price vs Transmition Type.

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = 'transmission', y = 'selling_price', data = df)
plt.title('Selling Price vs Transmition Type')
plt.xlabel('Transmition Type')
plt.ylabel('Selling Price')
plt.show()
plt.savefig('selling_price_vs_transmission_type.png',dpi=500,bbox_inches='tight')
```

- It shows Automatic vehicles have more selling price.

## Selling Price vs Ownership Type

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = 'owner', y = 'selling_price', data = df)
plt.title('Selling Price vs Ownership Type')
plt.xlabel('Ownership Type')
plt.xticks(rotation = 45)
plt.ylabel('Selling Price')
plt.show()
plt.savefig('selling_price_vs_ownership_type.png',dpi=500,bbox_inches='tight')
```

## Selling Price vs Seller Type

## Correlation Heatmap (Numeric)


```python
plt.figure(figsize = (10, 6))
sns.heatmap(data = df.select_dtypes(include=np.number).corr(), annot = True, cmap = 'coolwarm')
plt.title('Correlation Heatmap')
plt.show()
plt.savefig('correlation_heatmap_on_numerical_datas.png',dpi=500,bbox_inches='tight')
```

- Selling Price and Year Have a positive correlation
- Selling Price and Kilometers driven have a moderate level negative correlation
- Sellinng Price and Maximum Power Produced by the vehicle have a strong positive correlation
- Maximum Power and Engine Capacity have strong correlation.

## Brand-Wise Average Price

```python
brand_price = df.groupby('brand')['selling_price'].mean().sort_values(ascending = False)

plt.figure(figsize = (10,5))
brand_price.plot(kind = 'bar')
plt.title('Brand Wise Average Selling Price')
plt.xlabel('Brand')
plt.ylabel('Selling Price')
plt.show()
plt.savefig('brand_wise_selling_price.png',dpi=500,bbox_inches='tight')
```

- Premium brands have higher average prices.
- Brand plays a severe role on vehicle price.

### Key Insights from EDA

- Selling price is strongly influenced by vehicle age and usage
- Diesel and automatic vehicles generally have higher resale values
- First-owner cars command premium prices
- Brand significantly impacts selling price
- Outliers represent genuine market segments rather than data errors
